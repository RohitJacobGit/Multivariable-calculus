```python
def function_1_custom_1(i):
    k = 0.3
    return - k * i

def solve_function_1():
    step_size = 0.5
    num_of_steps = 40
    t = np.zeros(num_of_steps + 1)
    x = np.zeros(num_of_steps + 1)
    
    for i in range(num_of_steps):
        x[0] = 5 # initial condition
        t[i + 1] = step_size * (i + 1)
        x[i + 1] = x[i] + step_size * function_1_custom_1(x[i])
        
    return t,x

t, p = solve_function_1()

plt.plot(t,p,marker='.')
plt.ylabel('${dy}/{dt} =  - k * y$')
plt.xlabel('$time$')
```

```python
def function_1_custom_2(x):
    k = 0.3
    return - k * x

x = []

def System(x_init, step_size, num_of_steps):
    x.append(x_init)
    for i in range(num_of_steps):
        x.append(x[i] + step_size * function_1_custom_2(x[i]))
    return x

x = System(5, 0.5, 39)
t = np.arange(0, 20, 0.5)

plt.plot(t,x,marker='.')
plt.ylabel('${dy}/{dt} =  - k * y$')
plt.xlabel('$time$')
```

```python
def function_1_odeint(y,t,k):
    dydt = -k * y
    return dydt

# initial condition
y0 = 5
k = 0.3

# time points
t = np.linspace(0,20)

# solve ODE
y = odeint(function_1_odeint,y0,t,args=(k,))

# plot results
plt.plot(t,y,marker='.')
plt.ylabel('${dy}/{dt} =  - k * y$')
plt.xlabel('$time$')
```

```python
def lotka_volterra_1(x, y):
    #a = 1.
    #b = 0.1 
    return x * (a - b * y)

def lotka_volterra_2(x, y):
    #c = 1.5
    #d = 0.75
    #b = 0.1
    return -y * ( c - d*b*x)
```

```python
def solve_lotka_volterra():
    step_size = 15/1000
    num_of_steps = 1000
    # t = np.linspace(0,15,1000)
    t = np.zeros(num_of_steps+1)
    x = np.zeros(num_of_steps+1)
    y = np.zeros(num_of_steps+1)
    
    for i in range(num_of_steps):
        x[0] = 10
        y[0] = 5
        t[i+1] = (step_size) * (i+1)
        x[i+1] = x[i] + step_size * lotka_volterra_1(x[i],y[i])
        y[i+1] = y[i] + step_size * lotka_volterra_2(x[i],y[i])
        
    return t,x,y
```

```python
t, x_s, y_s = solve_lotka_volterra()
```

```python
plt.grid()
plt.xlabel('time')
plt.ylabel('population')
plt.plot(t,x_s,color='r',label='rabit')
plt.plot(t,y_s,color='b',label='foxes')
plt.legend(loc='best')
```
