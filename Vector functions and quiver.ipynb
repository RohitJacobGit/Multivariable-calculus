$f(x,y) = x^{2} + y^{2}$

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
plt.contour(x,y,z)
plt.grid(True)
plt.axis('scaled')
```

$f(x,y) = x^{2} + y^{2} = z = 2,4,6,8...$

```python
cs = plt.contour(x,y,z,levels,cmap='jet')
plt.clabel(cs, inline=True,fontsize=10)
```

$f(x,y) = 10 => x^{2}+y^{2} = 8$ 

```python
levels = [8]
cs = plt.contour(x,y,z,levels)
plt.clabel(cs, inline=True,fontsize=10)
```

## *******************************
## $f(x,y) = x + y$
## *******************************

```python
x = np.arange(-2,2,0.1)
y = np.arange(-2,2,0.1)
x, y = np.meshgrid(x,y)
z = x + y

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x+y$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y) 
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(10, 8))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(8, 8))  
ax = fig.add_subplot(111)  
ax.contourf(X, Y , Z, cmap='rainbow')
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)
plt.show()
```

## *******************************
## $f(x,y) = sin(x+y)$
## *******************************

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = np.arange(-np.pi, np.pi, 0.1)
x, y = np.meshgrid(x, y)
z = np.sin(x + y)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y ,z, cmap='jet', rstride=1, cstride=1)
ax.set_title("$f(x,y)= sin(x+y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

## *******************************
## $f(x,y) = sin(x)cos(y)$
## *******************************

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = np.arange(-np.pi, np.pi, 0.1)
x, y = np.meshgrid(x, y)
z = np.sin(x) * np.cos(y)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y ,z, cmap='jet', rstride=1, cstride=1)
ax.set_title("$f(x,y)= sin(x)cos(y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

## *******************************
## $f(x,y) = e^{(x + y)}$

```python
x = np.arange(-2, 2, 0.1)
y = np.arange(-2, 2, 0.1)
x, y = np.meshgrid(x, y)
z = np.exp(x+y)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y ,z, cmap='rainbow', rstride=1, cstride=1)
ax.set_title("$f(x,y)= e^{(x + y)}$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

## *******************************
## $f(t) = \begin{bmatrix} cos(t) \\ tsin(t) \end{bmatrix}$
## *******************************

```python
t = np.arange(0, 4*np.pi, 0.1)
X = np.cos(t)
Y = t * np.sin(t)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.set_title("$f(t)=[cos(t) \ tsin(t)]$", fontsize=18)
ax.plot(X, Y, color='r')
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.axis('equal')
plt.show()
```

## *******************************
## $f(t,s)=\begin{bmatrix}3cos(t)+cost(t)cos(s) \\\    3sin(t)+sin(t)cos(s) \\\ sin(s)\end{bmatrix}$
## *******************************

```python
t = np.linspace(-np.pi,np.pi,100)
s = np.linspace(-np.pi,np.pi,100)
t, s = np.meshgrid(t, s)
X = 3 * np.cos(t) + np.cos(t)*np.cos(s)
Y = 3 * np.sin(t) + np.sin(t)*np.cos(s)
Z = np.sin(s)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.set_title("$f(t,s)=[3cos(t)+cost(t)cos(s)$ , $3sin(t)+sin(t)cos(s)$ , $sin(s)]$", fontsize=18)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$x_3$", fontsize=18)
ax.set_zlim(-2, 2)
plt.show()
```

# Vector Fields

## *******************************

## $ f(x,y) = \begin{bmatrix} x \\ y\end{bmatrix} $

## *******************************

```python
X = np.linspace(-10, 10, 10)
Y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(X, Y)
X1 = X
X2 = Y

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.quiver(X,Y,X1,X2,edgecolor='b', facecolor='r', linewidth=0.2)
plt.show()
```

$f(x,y)=\bigg[\begin{matrix}y^3 - 9y \\ x^3-9x \end{matrix}\bigg]$

```python
X = np.linspace(-10, 10, 10)
Y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(X, Y)
X1 = Y**3 - 9 * Y
X2 = X**3 - 9 * X

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.quiver(X,Y,X1,X2,edgecolor='b', facecolor='r', linewidth=0.2)
plt.show()
```

$f(x,y,z)=\begin{bmatrix}yz \\ xz \\ xy\end{bmatrix}$

```python
X = np.linspace(-5, 5, 10)
Y = np.linspace(-5, 5, 10)
Z = np.linspace(-5, 5, 10)
X, Y, Z = np.meshgrid(X, Y, Z)
X1 = Y * Z
X2 = X * Z
X3 = X * Y

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111,projection='3d')
ax.quiver(X,Y,Z,X1,X2,X3, length=0.05)
#ax.quiver(X,Y,Z,X1,X2,X3,edgecolor='b', facecolor='r', linewidth=0.2, length=0.05)
plt.show()
```
